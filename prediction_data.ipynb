## Creating a Test Data Frame from Recent Historical Data Not Used In Training
Copying steps from data cleaning and removing fights used for training

```python
event_details = pd.read_csv("data_2023_10_14/ufc_event_details.csv")
fight_details = pd.read_csv("data_2023_10_14/ufc_fight_details.csv")
fight_results = pd.read_csv("data_2023_10_14/ufc_fight_results.csv")
fight_stats = pd.read_csv("data_2023_10_14/ufc_fight_stats.csv")
fighter_details = pd.read_csv("data_2023_10_14/ufc_fighter_details.csv")
fighter_tott = pd.read_csv("data_2023_10_14/ufc_fighter_tott.csv")
```

```python
# remove these fights from fight_results
cutoff_index = np.where(fight_results["BOUT"]=="Bobby Hoffman  vs. Mark Robinson ")[0][0]
fight_results_modern = fight_results[fight_results.index < cutoff_index].copy()

# remove these fights from fight_stats
cutoff_index = np.where(fight_stats["BOUT"]=="Bobby Hoffman vs. Mark Robinson")[0][0]
fight_stats_modern = fight_stats[fight_stats.index < cutoff_index].copy()
```

```python
event_details.drop(['URL', 'LOCATION'], axis=1, inplace=True)
fight_results_modern.drop(['REFEREE', 'DETAILS', 'URL'], axis=1, inplace=True)
fight_stats_modern.drop(['SIG.STR. %', 'TD %'], axis=1, inplace=True)
fighter_tott.drop(['URL', 'WEIGHT'], axis=1, inplace=True)
```

```python
fighter_tott['STANCE'] = fighter_tott['STANCE'].fillna("--")
```

```python
dfs = [event_details, fight_results_modern, fight_stats_modern, fighter_tott]

# function from https://www.geeksforgeeks.org/pandas-strip-whitespace-from-entire-dataframe/
# and modified to include removal of multiple spaces in sequence.
def remove_whitespace(dataframe):
   
    # iterating over the columns
    for i in dataframe.columns:
         
        # checking if column includes strings
        if dataframe[i].dtype == 'object':
            # applying strip function on column to remove leading and trailing white space
            dataframe[i] = dataframe[i].map(str.strip)
            # applying replace function on column to remove multiple spaces in sequence
            dataframe[i] = dataframe[i].str.replace(r'\s+', ' ', regex=True)
    
for df in dfs:
    remove_whitespace(df)
```

```python
is_title_bout = lambda x: ("Ultimate Fighter" not in x) and ("TUF" not in x) and ("Title" in x)
fight_results_modern['TITLE'] = fight_results_modern['WEIGHTCLASS'].apply(is_title_bout)

divisions = ["Flyweight", "Bantamweight", "Featherweight", "Lightweight", "Welterweight", "Middleweight", 
             "Light Heavyweight", "Heavyweight", "Women's Strawweight", "Women's Flyweight Bout", 
             "Women's Bantamweight Bout", "Women's Featherweight", "Catch Weight"]
pattern = '(' + '|'.join(divisions) + ')'

fight_results_modern['WEIGHTCLASS'] = fight_results_modern['WEIGHTCLASS'].str.extract(pattern)
```

```python
fight_results_modern['OUTCOME'] = fight_results_modern['OUTCOME'].replace({'W/L': 'R', 'L/W': 'B', 'NC/NC': 'NC', 'D/D': 'D'})
```

```python
fight_results_modern[['r_fighter', 'b_fighter']] = fight_results_modern['BOUT'].str.split(" vs. ", expand = True)
```

```python
def time_to_seconds(time_str):
    minutes, seconds = time_str.split(':')
    return int(minutes) * 60 + int(seconds)

fight_results_modern['tot_time_s'] = (fight_results_modern['ROUND'] - 1) * 300 + fight_results_modern['TIME'].apply(time_to_seconds)

# drop time because it is redundant to have both time in minutes and seconds.
fight_results_modern.drop(['TIME'], axis=1, inplace=True)
```

```python
fight_stats_modern['ctrl_s'] = fight_stats_modern['CTRL'].apply(time_to_seconds)
fight_stats_modern.drop(['CTRL'], axis=1, inplace=True)
```

```python
fight_stats_modern[['sig_str_landed', 'sig_str_att']] = fight_stats_modern['SIG.STR.'].str.split(" of ", expand = True)
fight_stats_modern[['total_str_landed', 'total_str_att']] = fight_stats_modern['TOTAL STR.'].str.split(" of ", expand = True)
fight_stats_modern[['td_landed', 'td_att']] = fight_stats_modern['TD'].str.split(" of ", expand = True)
fight_stats_modern[['head_landed', 'head_att']] = fight_stats_modern['HEAD'].str.split(" of ", expand = True)
fight_stats_modern[['body_landed', 'body_att']] = fight_stats_modern['BODY'].str.split(" of ", expand = True)
fight_stats_modern[['leg_landed', 'leg_att']] = fight_stats_modern['LEG'].str.split(" of ", expand = True)
fight_stats_modern[['distance_landed', 'distance_att']] = fight_stats_modern['DISTANCE'].str.split(" of ", expand = True)
fight_stats_modern[['clinch_landed', 'clinch_att']] = fight_stats_modern['CLINCH'].str.split(" of ", expand = True)
fight_stats_modern[['ground_landed', 'ground_att']] = fight_stats_modern['GROUND'].str.split(" of ", expand = True)
fight_stats_modern.drop(['SIG.STR.', 'TOTAL STR.', 'TD', 'HEAD', 'BODY', 'LEG', 'DISTANCE', 'CLINCH', 'GROUND'], axis=1, inplace=True)
```

```python
fight_stats_modern['ROUND'] = fight_stats_modern['ROUND'].str.extract('(\d+)').astype(int)
fighter_tott['HEIGHT'] = fighter_tott['HEIGHT'].apply(lambda x: int(x.split("'")[0])*12 + int(x.split("'")[1][:-1]) if x != "--" else np.nan)
fighter_tott['REACH'] = fighter_tott['REACH'].apply(lambda x: int(x[:-1]) if x != "--" else np.nan)
```

```python
def convert_object(df):
    for col in df.select_dtypes(include='object').columns:
        try:
            df[col] = df[col].astype('int')
        except ValueError:
            df[col] = df[col].astype('string')
    return df

event_details = convert_object(event_details)

event_details['DATE'] = pd.to_datetime(event_details['DATE']).astype(int)/10**11

fight_results_modern = convert_object(fight_results_modern)
fight_stats_modern = convert_object(fight_stats_modern)
for col in fight_stats_modern.select_dtypes(include='float').columns:
        fight_stats_modern[col] = fight_stats_modern[col].astype('int')
fighter_tott = convert_object(fighter_tott)

def convert_to_int(date):
    if date == '--':
        return np.nan
    try:
        return int(pd.to_datetime(date).timestamp())/10**2
    except ValueError:
        return np.nan

fighter_tott['DOB'] = fighter_tott['DOB'].apply(convert_to_int)
```

```python
fight_stats_modern_tott = fight_stats_modern.groupby(['EVENT', 'BOUT', 'FIGHTER'], sort=False).sum().reset_index()
fight_stats_modern_tott.drop(['ROUND'], axis=1, inplace=True)

from tqdm.notebook import tqdm_notebook
from statistics import mean

stats_dict = {}

for name in fight_stats_modern_tott['FIGHTER'].unique(): # for all fighters
    stats_dict[name] = {}
    for i in fight_stats_modern_tott.select_dtypes(include=['number']).columns: # for all numerical stats
        stats_dict[name].setdefault(i, []) #initialize a dictionary of fighter stats to empty array

agg_fight_stats = fight_stats_modern_tott.copy()

for i, row in tqdm_notebook(fight_stats_modern_tott[::-1].iterrows(), total=len(fight_stats_modern_tott[::-1])):
    for col in agg_fight_stats.select_dtypes(include=['number']).columns:
        if len(stats_dict[row['FIGHTER']][col]) == 0:
            agg_fight_stats.loc[(agg_fight_stats['EVENT'] == row['EVENT']) &
                                (agg_fight_stats['BOUT'] == row['BOUT']) &
                                (agg_fight_stats['FIGHTER'] == row['FIGHTER']), col] = np.nan
            continue
        else:
            agg_fight_stats.loc[(agg_fight_stats['EVENT'] == row['EVENT']) &
                                (agg_fight_stats['BOUT'] == row['BOUT']) &
                                (agg_fight_stats['FIGHTER'] == row['FIGHTER']), col] = mean(stats_dict[row['FIGHTER']][col])
    for col in agg_fight_stats.select_dtypes(include=['number']).columns:
            stats_dict[row['FIGHTER']][col].append(row[col])
    
# rename columns of agg_fight_stats
numeric_cols = agg_fight_stats.select_dtypes(include=['number']).columns
agg_fight_stats[numeric_cols] = agg_fight_stats[numeric_cols].add_prefix('avg_')
```

```python
# Change _landed stats to percentage of attempts landed to mitigate collinearity issues.
landed_cols = [col for col in agg_fight_stats.columns if col.endswith('landed')]

# Loop through the 'landed' columns and calculate the corresponding percentages
for col in landed_cols:
    att_col = col.replace('_landed', '_att')
    agg_fight_stats[col] = np.where(agg_fight_stats[att_col] == 0, 0, agg_fight_stats[col] / agg_fight_stats[att_col])

    
# Rename columns to reflect this change
agg_fight_stats.columns = [col.replace('_landed', '_percent') for col in agg_fight_stats.columns]
```

```python
# adding average time of fights
agg_fight_stats['avg_time_s'] = np.nan

time_dict = {}
for name in fight_stats_modern['FIGHTER'].unique(): # for all fighters
    time_dict[name] = []

for i, row in tqdm_notebook(fight_results_modern[::-1].iterrows(), total=len(fight_results_modern[::-1])):
    # update info for r_fighter
    if len(time_dict[row['r_fighter']]) != 0:
        agg_fight_stats.loc[(agg_fight_stats['EVENT'] == row['EVENT']) &
                                (agg_fight_stats['BOUT'] == row['BOUT']) & 
                                (agg_fight_stats['FIGHTER']  == row['r_fighter']), 'avg_time_s'] = mean(time_dict[row['r_fighter']])
    time_dict[row['r_fighter']].append(row['tot_time_s'])
    # update info for b_fighter
    if len(time_dict[row['b_fighter']]) != 0:
        agg_fight_stats.loc[(agg_fight_stats['EVENT'] == row['EVENT']) &
                                (agg_fight_stats['BOUT'] == row['BOUT']) & 
                                (agg_fight_stats['FIGHTER']  == row['b_fighter']), 'avg_time_s'] = mean(time_dict[row['b_fighter']])
    time_dict[row['b_fighter']].append(row['tot_time_s'])
```

```python
# Adding wins, losses, draws, win/loss streak, longest win streak, 
#     percent of bouts that are title bouts, percent of wins/losses by each method to agg_fight_stats
# This is done seperately as it deals with fight_results as opposed to fight_stats
results_columns = ['wins', 'losses', 'draws', 'win_streak', 'loss_streak', 'longest_win_streak', 'percent_title_bouts',
           'percent_win_by_ko/tko', 'percent_win_by_sub', 'percent_win_by_dec', 'percent_win_by_doctor_stoppage',
           'percent_loss_by_ko/tko', 'percent_loss_by_sub', 'percent_loss_by_dec', 'percent_loss_by_doctor_stoppage']

results_dict_keys = ['wins', 'losses', 'draws', 'win_streak', 'loss_streak', 'longest_win_streak', 'title_bouts',
                     'wins_by_ko/tko', 'wins_by_sub', 'wins_by_dec', 'wins_by_doctor_stoppage',
                     'losses_by_ko/tko', 'losses_by_sub', 'losses_by_dec', 'losses_by_doctor_stoppage']

for col in results_columns:
    agg_fight_stats[col] = 0
    
results_dict = {}
for name in fight_stats_modern_tott['FIGHTER'].unique(): # for all fighters
    results_dict[name] = {}
    for i in results_dict_keys:
        results_dict[name].setdefault(i, 0) #initialize a dictionary of fighter results to 0's

for i, row in tqdm_notebook(fight_results_modern[::-1].iterrows(), total=len(fight_results_modern[::-1])):
    # ---- adding results to agg_fight_stats ----
    # for r_fighter
    # add wins, losses, draws, win_streak, loss_streak, and longest_win_streak to agg_fight_stats
    for i in range(6):
        agg_fight_stats.loc[(agg_fight_stats['EVENT'] == row['EVENT']) &
                            (agg_fight_stats['BOUT'] == row['BOUT']) & 
                            (agg_fight_stats['FIGHTER']  == row['r_fighter']), results_columns[i]] \
        = results_dict[row['r_fighter']][results_dict_keys[i]]
    # add percent_title_bouts to agg_fight_stats
    if (results_dict[row['r_fighter']]['wins'] + results_dict[row['r_fighter']]['losses'] + 
         results_dict[row['r_fighter']]['draws']) > 0:
        agg_fight_stats.loc[(agg_fight_stats['EVENT'] == row['EVENT']) &
                            (agg_fight_stats['BOUT'] == row['BOUT']) & 
                            (agg_fight_stats['FIGHTER']  == row['r_fighter']), 'percent_title_bouts'] \
        = results_dict[row['r_fighter']]['title_bouts'] / (results_dict[row['r_fighter']]['wins']
                                                           + results_dict[row['r_fighter']]['losses']
                                                           + results_dict[row['r_fighter']]['draws'])
    # add percentage of wins columns to agg_fight_stats
    if results_dict[row['r_fighter']]['wins'] > 0:
        for i in range(7, 11):
            agg_fight_stats.loc[(agg_fight_stats['EVENT'] == row['EVENT']) &
                                (agg_fight_stats['BOUT'] == row['BOUT']) & 
                                (agg_fight_stats['FIGHTER']  == row['r_fighter']), results_columns[i]] \
            = results_dict[row['r_fighter']][results_dict_keys[i]]/results_dict[row['r_fighter']]['wins']
    # add percentage of losses columns to agg_fight_stats
    if results_dict[row['r_fighter']]['losses'] > 0:
        for i in range(11, 15):
            agg_fight_stats.loc[(agg_fight_stats['EVENT'] == row['EVENT']) &
                                (agg_fight_stats['BOUT'] == row['BOUT']) & 
                                (agg_fight_stats['FIGHTER']  == row['r_fighter']), results_columns[i]] \
            = results_dict[row['r_fighter']][results_dict_keys[i]]/results_dict[row['r_fighter']]['losses']

    # -------------------------------------------
    
    # ---------- updating results_dict ----------
    if row['TITLE'] == True: #if bout was for a title bout, increase the number of title bouts for both fighters.
        results_dict[row['r_fighter']]['title_bouts'] += 1
        results_dict[row['b_fighter']]['title_bouts'] += 1
        
    if row['OUTCOME'] == "NC" or row['METHOD'] == "D": #if result is a draw, increment draws for both fighters.
        results_dict[row['r_fighter']]['draws'] += 1
        results_dict[row['b_fighter']]['draws'] += 1
    #if result is not a draw, increment win/loss stats appropriately
    elif row['OUTCOME'] == "R":
        results_dict[row['r_fighter']]['wins'] += 1
        results_dict[row['b_fighter']]['losses'] += 1
        
        if row['METHOD'] == "KO/TKO":
            results_dict[row['r_fighter']]['wins_by_ko/tko'] += 1
            results_dict[row['b_fighter']]['losses_by_ko/tko'] += 1
        elif row['METHOD'] == "Submission":
            results_dict[row['r_fighter']]['wins_by_sub'] += 1
            results_dict[row['b_fighter']]['losses_by_sub'] += 1
        elif row['METHOD'] in ["Decision - Unanimous", "Decision - Split", "Decision - Majority"]:
            results_dict[row['r_fighter']]['wins_by_dec'] += 1
            results_dict[row['b_fighter']]['losses_by_dec'] += 1
        elif row['METHOD'] == "TKO - Doctor's Stoppage":
            results_dict[row['r_fighter']]['wins_by_doctor_stoppage'] += 1
            results_dict[row['b_fighter']]['losses_by_doctor_stoppage'] += 1
        
        results_dict[row['r_fighter']]['win_streak'] += 1
        results_dict[row['b_fighter']]['loss_streak'] += 1
        if results_dict[row['r_fighter']]['win_streak'] > results_dict[row['r_fighter']]['longest_win_streak']:
            results_dict[row['r_fighter']]['longest_win_streak'] = results_dict[row['r_fighter']]['win_streak']
        if results_dict[row['r_fighter']]['loss_streak'] != 0: results_dict[row['r_fighter']]['loss_streak'] == 0   
        if results_dict[row['b_fighter']]['win_streak'] != 0: results_dict[row['b_fighter']]['win_streak'] == 0
            
    elif row['OUTCOME'] == "B":
        results_dict[row['r_fighter']]['losses'] += 1
        results_dict[row['b_fighter']]['wins'] += 1
        if row['METHOD'] == "KO/TKO":
            results_dict[row['r_fighter']]['losses_by_ko/tko'] += 1
            results_dict[row['b_fighter']]['wins_by_ko/tko'] += 1
        elif row['METHOD'] == "Submission":
            results_dict[row['r_fighter']]['losses_by_sub'] += 1
            results_dict[row['b_fighter']]['wins_by_sub'] += 1
        elif row['METHOD'] in ["Decision - Unanimous", "Decision - Split", "Decision - Majority"]:
            results_dict[row['r_fighter']]['losses_by_dec'] += 1
            results_dict[row['b_fighter']]['wins_by_dec'] += 1
        elif row['METHOD'] == "TKO - Doctor's Stoppage":
            results_dict[row['r_fighter']]['losses_by_doctor_stoppage'] += 1
            results_dict[row['b_fighter']]['wins_by_doctor_stoppage'] += 1
            
        results_dict[row['b_fighter']]['win_streak'] += 1
        results_dict[row['r_fighter']]['loss_streak'] += 1
        if results_dict[row['b_fighter']]['win_streak'] > results_dict[row['b_fighter']]['longest_win_streak']:
            results_dict[row['b_fighter']]['longest_win_streak'] = results_dict[row['b_fighter']]['win_streak']
        if results_dict[row['b_fighter']]['loss_streak'] != 0: results_dict[row['b_fighter']]['loss_streak'] == 0
        if results_dict[row['r_fighter']]['win_streak'] != 0: results_dict[row['r_fighter']]['win_streak'] == 0
    # -------------------------------------------
```
